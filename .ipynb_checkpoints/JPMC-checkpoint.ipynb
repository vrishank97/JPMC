```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train = pd.read_csv("housing_train.csv")
test = pd.read_csv("housing_test.csv")
```

```python
train.head()
```

```python
test.head()
```

```python
from sklearn.preprocessing import LabelEncoder

categorical=[]
for c in train.columns:
    if train[c].dtype == 'object':
        categorical.append(c)

for c in train.columns:
    if train[c].dtype == 'object':
        lbl = LabelEncoder() 
        lbl.fit(list(train[c].values) + list(test[c].values)) 
        train[c] = lbl.transform(list(train[c].values))
        test[c] = lbl.transform(list(test[c].values))
```

```python
print('Shape train: {}\nShape test: {}'.format(train.shape, test.shape))
```

```python
train = train.drop(["TID"], axis=1)
test = test.drop(["TID"], axis=1)
```

```python
y = train["Price"]
```

```python
import numpy as np
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=50, n_jobs=-1)

model.fit(train.drop(["Price"], axis=1), np.log(y))
pred = model.predict(test)
pred = np.exp(pred)
with open('your_file8.txt', 'w') as f:
    for item in pred:
        f.write("%s, " % item)
```

```python
model = RandomForestRegressor(n_estimators=100, n_jobs=-1)

model.fit(train.drop(["Price"], axis=1), np.log(y))
pred = model.predict(test)
pred = np.exp(pred)
with open('your_file9.txt', 'w') as f:
    for item in pred:
        f.write("%s, " % item)
```
